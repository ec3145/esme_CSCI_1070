```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
survey_df = pd.read_csv("CSCI_Fall_2024_survey.csv")
survey_df
```

```python
survey_df.drop(["What state/country were you born in?"],inplace=True, axis=1)
survey_df = survey_df.assign(Country = ['Mexico','US','Germany','US','US','US','India','US','US','Georgia',
                                      'US','US','US','US','US','US','US','US','US','US','India','Peoples Republic of China',
                                      'India','Cameroon','US','Kazakhstah'])
survey_df
```

```python
country_state = survey_df.groupby(survey_df['Country']).count()
country_state
```

```python
colors = ['pink','green']
country_state.plot(kind="pie", y = "What is your name?", colors=colors)
# this pie chart shows that mostly everyone is from the US but there are still quite a bit of people that were born in other places. 
```

```python
survey_df['Dogs'] = np.where(survey_df["How many pets and of which kind do you have?"]
                            .str.contains('dog'),1,
                             np.where(survey_df["How many pets and of which kind do you have?"]
                            .str.contains('Dog'),1,
                                      np.where(survey_df["How many pets and of which kind do you have?"]
                                               .str.contains('schnauzers'),1,0)))
survey_df['Cats'] = np.where(survey_df["How many pets and of which kind do you have?"]
                            .str.contains('cat'),1,
                             np.where(survey_df["How many pets and of which kind do you have?"]
                            .str.contains('Cats'),1,0))
survey_df['None'] = np.where(survey_df["How many pets and of which kind do you have?"]
                            .str.contains('none'),1,
                             np.where(survey_df["How many pets and of which kind do you have?"]
                            .str.contains('0'),1,
                                      np.where(survey_df["How many pets and of which kind do you have?"]
                                               .str.contains('no'),1,0)))
survey_df
```

```python
plt.plot(survey_df["Dogs"],survey_df["Cats"],survey_df["None"],'.')
#this scatter plot was supposed to show who has a pet and who doesnt but it didnt really end up shwoing that. 
```

```python
survey_df.drop(["What is your favorite type of food?"],inplace=True, axis=1)
survey_df = survey_df.assign(Fav_Food = ['J','M','ITLN','ITLN','A','J','IDN','M','C','M','A','ITLN',
                                      'IDN','ITLN','M','IDN','J','ITLN','ALL','A','ITLN','J','J','C',
                                      'K','M'])
survey_df
```

```python
food=survey_df['Fav_Food'].value_counts()
food.plot(kind="bar")
plt.xlabel('Favorite Foods')
plt.ylabel('Count')
plt.title('Histogram of Favorite Food')
#this histogram shows that students favorite food is italian but also japan and mexican are tied.
```

```python
survey_df.drop(["What is your major?"],inplace=True, axis=1)
survey_df = survey_df.assign(major = ['CS','CS/BIZZ','DSCIENCE','BIZZ ECON','CS','CS','CS/fiN','CS','CS','DSCIENCE','CS','CS',
                                      'CS','CS','CS','CS','BIO','CS','DSCIENCE','CS','CS','TBD','CS','CS',
                                      'ECON','DSCIENCE'])
survey_df
```

```python
food=survey_df['major'].value_counts()
food.plot(kind="bar")
plt.xlabel('Majors')
plt.ylabel('Count')
plt.title('Histogram of majors')
```

```python
survey_df.drop(["How old are you?"],inplace=True, axis=1)
survey_df = survey_df.assign(age = ['Teen','Teen','Teen','Adult','Teen','Teen','Teen','Teen','Teen','Teen','Teen','Teen',
                                      'Teen','Teen','Teen','Teen','Adult','Teen','Teen','Teen','Teen','Adult','Teen','Teen',
                                      'Teen','Teen'])
survey_df
```

```python
teen_adult = survey_df.groupby("age").count()
teen_adult
```

```python
colors = ['green','blue']
teen_adult.plot(kind="pie", y = "What is your name?", colors=colors)
#in this pie chart it showcases who is an adult and who is considered a "teen" based on the teen at the end of the word.
#it shows there are more students that are "teens" than adults.
```
