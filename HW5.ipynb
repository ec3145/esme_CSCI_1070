```python
import pandas as pd
import numpy as np

titanic_df = pd.read_csv("titanic.csv")
titanic_df
```

```python
titanic_df.columns
```

One-hot Encoding 

```python
titanic_df = pd.get_dummies(titanic_df[['Sex']])
titanic_df
```

categorical to numerical

```python
embark_map = {'S':1,'C':2,'Q':3}
titanic_df['Embarked'] = titanic_df['Embarked'].map(embark_map)
titanic_df
```

cleaning null values

```python
titanic_df.fillna('')
```
